```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data collection and Proceessing

```python
#loading the data from sklearn

breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
#loading the data to the data frame
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
x=data_frame.drop(columns='label',axis=1)
y=data_frame['label']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Standrdise the data


```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train_std=scaler.fit_transform(x_train)
x_test_std=scaler.transform(x_test)
```

Building the neural network

```python
# importing tensor flow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.Input(shape=(30,)),
    keras.layers.Flatten(),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

0=Malignent
1=benignent

```python
history = model.fit(
   x_train_std,
    y_train,
    validation_split=0.1,
    epochs=10
)
```

Visulize accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')

plt.legend(['Training Data', 'Validation Data'], loc='lower right')
plt.show()
```

Accuracy of model on test data

```python
loss, accuracy=model.evaluate(x_test_std,y_test)
print(accuracy)
```

Building the predcitive system

```python
import numpy as np

input_data = (20.57,17.77,132.9,1326,0.08474,0.07864,0.0869,0.07017,0.1812,
              0.05667,0.5435,0.7339,3.398,74.08,0.005225,0.01308,0.0186,
              0.0134,0.01389,0.003532,24.99,23.41,158.8,1956,0.1238,
              0.1866,0.2416,0.186,0.275,0.08902)

input_data_np = np.asarray(input_data).reshape(1,-1)

# scale input
input_data_std = scaler.transform(input_data_np)

# prediction
prediction = model.predict(input_data_std)
print(prediction)

# convert probability → class
prediction_label = (prediction > 0.5).astype(int)
print(prediction_label)

if prediction_label[0][0] == 0:
    print('The breast cancer is Malignant')
else:
    print('The breast cancer is Benign')
```
